# CTAB to JSON

Taking gtmseg.ctab (FreeSurfer's expanded color table) and converting it to JSON dictionary.

```python
import pandas as pd
import json
import numpy as np
```

```python
fields = ['region_index','name','color_r','color_g','color_b','alpha','tissue_type']
ctab = pd.read_csv('gtmseg.ctab',sep='\s+',comment='#',names=fields)
```

```python
ctab
```

```python
ctab_json = ctab[['region_index','name']]
```

```python
ctab_json
```

```python
ctab_json.to_json('gtmseg_ctab.json',index=False,orient='split',indent=4)
```

```python
new_ctab = pd.read_json('gtmseg_ctab.json',orient='split')
```
